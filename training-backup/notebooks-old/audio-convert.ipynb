This notebook converts all dataset audio into a new directory. Requires FFMPEG installed.

```python
from os import path

SOURCE_DIRECTORY = path.normpath("../datasets/opus")
TARGET_DIRECTORY = path.normpath("../datasets/pcm_s16le_441")
TARGET_EXTENSION = "wav"
TARGET_CODEC = "pcm_s16le"
TARGET_SAMPLE_RATE = 44_100 # Hz
```

```python
import subprocess
from os import listdir, makedirs, path, remove
from tqdm.notebook import tqdm

def target_path_for(src_like_path):
    base, _ = path.splitext(src_like_path)
    return f"{base}.{TARGET_EXTENSION}"

def is_valid_file(src_path):
    if not path.isfile(src_path):
        return False

    result = subprocess.run(
        [
            "ffprobe",
            "-hide_banner",
            "-v",
            "error",
            "-select_streams",
            "a:0",
            "-show_entries",
            "stream=codec_name,sample_rate",
            "-of",
            "default=noprint_wrappers=1",
            src_path,
        ],
        capture_output=True,
        text=True,
    )
    output = result.stdout
    return TARGET_CODEC in output and str(TARGET_SAMPLE_RATE) in output


def convert_file(src_path, dst_path):
    final_dst_path = target_path_for(dst_path)
    cmd = [
        "ffmpeg",
        "-hide_banner",
        "-y",
        "-i",
        src_path,
        "-ar",
        str(TARGET_SAMPLE_RATE),
        "-ac",
        "1",
        "-c:a",
        TARGET_CODEC,
        final_dst_path,
    ]
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.returncode != 0:
        if path.isfile(final_dst_path):
            remove(final_dst_path)
        return ("failed", result.stderr)
    return "converted"


converted = 0
skipped = 0
failed = 0

makedirs(TARGET_DIRECTORY, exist_ok=True)

for dataset_type in tqdm(listdir(SOURCE_DIRECTORY), desc="Processing datasets"):
    type_path = path.join(SOURCE_DIRECTORY, dataset_type)
    if not path.isdir(type_path):
        continue

    makedirs(path.join(TARGET_DIRECTORY, dataset_type), exist_ok=True)

    for class_name in tqdm(listdir(type_path), desc=f"Processing {dataset_type} classes"):
        class_path = path.join(type_path, class_name)
        if not path.isdir(class_path):
            continue

        makedirs(path.join(TARGET_DIRECTORY, dataset_type, class_name), exist_ok=True)

        for audio_file in tqdm(listdir(class_path), desc=f"Processing {class_name}"):
            audio_path = path.join(class_path, audio_file)
            if not path.isfile(audio_path):
                continue

            target_audio_path = path.join(TARGET_DIRECTORY, dataset_type, class_name, audio_file)
            target_converted_path = target_path_for(target_audio_path)

            if is_valid_file(target_converted_path):
                skipped += 1
                continue

            status = convert_file(audio_path, target_audio_path)

            if status == "converted":
                converted += 1
            else:
                failed += 1
                print(f"[FAILED] {audio_path}: {status[1]}")

print(f"\nConverted: {converted} | Skipped: {skipped} | Failed: {failed}")
```
