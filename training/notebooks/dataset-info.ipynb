```python
from os import path, listdir

SOURCE_PATH      = path.normpath("../datasets/opus")
AUDIO_SAMPLE_RATE = 48_000
```

```python
import librosa
import pandas as pd
from tqdm.notebook import tqdm
import warnings

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

records = []  # list of dicts: {dataset_type, class_name, filename, duration_s}

for dataset_type in tqdm(listdir(SOURCE_PATH), desc="Dataset types"):
    type_path = path.join(SOURCE_PATH, dataset_type)
    if not path.isdir(type_path):
        continue

    for class_name in tqdm(listdir(type_path), desc=f"  {dataset_type}", leave=False):
        class_path = path.join(type_path, class_name)
        if not path.isdir(class_path):
            continue

        for audio_file in tqdm(listdir(class_path), desc=f"   {dataset_type} - {class_name}"):
            audio_path = path.join(class_path, audio_file)
            if not path.isfile(audio_path):
                continue

            try:
                duration = librosa.get_duration(path=audio_path)
                records.append({
                    "dataset_type": dataset_type,
                    "class_name":   class_name,
                    "filename":     audio_file,
                    "duration_s":   duration,
                })
            except Exception as e:
                print(f"  [WARN] Could not read {audio_path}: {e}")
```

```python
df = pd.DataFrame(records)
print(f"\nTotal audio files found : {len(df)}")
df.groupby(["dataset_type", "class_name"])["duration_s"].agg(count="count", min="min", mean="mean", max="max").round(5)
```

```python
import matplotlib.pyplot as plt

# ── 2. Helper – consistent box-plot style ─────────────────────────────────────
BOXPROPS   = dict(color="#01696f", linewidth=1.4)
WHISKERPROPS = dict(color="#01696f", linewidth=1.2, linestyle="--")
MEDIANPROPS  = dict(color="#da7101", linewidth=2)
FLIERPROPS   = dict(marker="o", markerfacecolor="#a12c7b", markersize=4,
                    linestyle="none", alpha=0.6)

def _base_fig(n_groups, title, xlabel):
    fig, ax = plt.subplots(figsize=(max(6, n_groups * 1.4), 5))
    ax.set_title(title, fontsize=13, fontweight="bold", pad=12)
    ax.set_xlabel(xlabel, fontsize=11)
    ax.set_ylabel("Duration (seconds)", fontsize=11)
    ax.grid(axis="y", linestyle="--", alpha=0.4)
    ax.spines[["top", "right"]].set_visible(False)
    return fig, ax

# ── 3. Plot A – one box per DATASET TYPE ──────────────────────────────────────
dataset_types = sorted(df["dataset_type"].unique())
data_by_type  = [df.loc[df["dataset_type"] == dt, "duration_s"].values
                 for dt in dataset_types]

fig, ax = _base_fig(len(dataset_types),
                    "Audio Duration Distribution by Dataset Type",
                    "Dataset Type")
ax.boxplot(data_by_type,
           tick_labels=dataset_types,
           boxprops=BOXPROPS, whiskerprops=WHISKERPROPS,
           medianprops=MEDIANPROPS, flierprops=FLIERPROPS,
           patch_artist=True)
plt.xticks(rotation=30, ha="right", fontsize=9)
plt.tight_layout()
plt.show()

# ── 4. Plot B – one box per CLASS, faceted by DATASET TYPE ────────────────────
for dt in dataset_types:
    subset      = df[df["dataset_type"] == dt]
    class_names = sorted(subset["class_name"].unique())
    data_by_cls = [subset.loc[subset["class_name"] == cn, "duration_s"].values
                   for cn in class_names]

    fig, ax = _base_fig(len(class_names),
                        f"Duration Distribution by Class  ·  [{dt}]",
                        "Class")
    ax.boxplot(data_by_cls,
               tick_labels=class_names,
               boxprops={**BOXPROPS, "facecolor": "#cedcd8"},
               whiskerprops=WHISKERPROPS,
               medianprops=MEDIANPROPS,
               flierprops=FLIERPROPS,
               patch_artist=True)
    plt.xticks(rotation=45, ha="right", fontsize=8)
    plt.tight_layout()
    plt.show()
```
