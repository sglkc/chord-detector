This notebook converts all dataset audio into a new directory. Requires FFMPEG installed.

```python
SOURCE_DIRECTORY = "../datasets/pcm_s32le"
TARGET_DIRECTORY = "../datasets/opus_pcm"
TARGET_CODEC = "libopus"
TARGET_SAMPLE_RATE = 48_000
```

```python
import subprocess
from os import makedirs, path, remove, listdir
from tqdm.notebook import tqdm

def convert_file(src_path, dst_path):
    cmd = [
        'ffmpeg', '-y', '-i', src_path,
        '-ar', str(TARGET_SAMPLE_RATE), '-ac', '1',
        '-c:a', TARGET_CODEC, dst_path.replace('wav', 'webm')
    ]
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.returncode != 0:
        if path.isfile(dst_path):
            remove(dst_path)
        return ('failed', result.stderr)
    return 'converted'

converted = 0
failed = 0

makedirs(path.dirname(TARGET_DIRECTORY), exist_ok=True)

for dataset_type in tqdm(listdir(SOURCE_DIRECTORY), desc="Processing datasets"):
  type_path = path.join(SOURCE_DIRECTORY, dataset_type)
  makedirs(path.join(TARGET_DIRECTORY, dataset_type), exist_ok=True)
  
  for class_name in tqdm(listdir(type_path), desc=f"Processing {dataset_type} classes"):
    class_path = path.join(type_path, class_name)
    makedirs(path.join(TARGET_DIRECTORY, dataset_type, class_name), exist_ok=True)

    if path.isdir(class_path):
      for audio_file in tqdm(listdir(class_path), desc=f"Processing {class_name}"):
        audio_path = path.join(class_path, audio_file)
        
        if path.isfile(audio_path):
          target_audio_path = path.join(TARGET_DIRECTORY, dataset_type, class_name, audio_file)
          status = convert_file(audio_path, target_audio_path)

          if status == "converted":
              converted += 1
          else:
              failed += 1
              print(f"[FAILED] {audio_path}: {status[1]}")

print(f"\nConverted: {converted} | Failed: {failed}")
```
